# Reading PDB data

### 1.) This time we'll read from a file instead of embedding the text in the notebook. We can simply draw the molecule with MolToImage function. 

```python
molblock_crn = Chem.MolFromPDBFile('1CRN.pdb', removeHs=True)
Draw.MolToImage(molblock_crn, size=(800, 800))
```

```python
from rdkit.Chem.Draw import rdMolDraw2D
from IPython.display import SVG

drawer = rdMolDraw2D.MolDraw2DSVG(1000, 1000)
drawer.DrawMolecule(molblock_crn)
drawer.FinishDrawing()

SVG(drawer.GetDrawingText())
```

### 2) Sanity check

```python
# To get number of atoms
molblock_crn is not None, molblock_crn.GetNumAtoms(), molblock_crn.GetNumBonds()
```

### 3) Look at the molecule "as text" (SMILES and formula)

```python
from rdkit.Chem import rdMolDescriptors # rdMolDescriptors is a module of the rdkit.Chem
smiles = Chem.MolToSmiles(molblock_crn) 
formula = rdMolDescriptors.CalcMolFormula(molblock_crn)
smiles[:120], formula
```

##### i) SMILES is RDKit's linear representation (for proteins it can be huge / not super meaningul, but still useful) 
##### ii) Molecular formula gives quick composition

### 4) Inspect atoms: element, index, neigbors, charge

```python
for a in molblock_crn.GetAtoms():
    print(a.GetIdx(), a.GetSymbol(), "deg=", a.GetDegree(), "charge=", a.GetFormalCharge())
```

### 5) Get 3D coordinates (PDB's main value)

```python
# PDBs are about 3D geometry. RDKit can access conformer coordinates

conf = molblock_crn.GetConformer()
for i in range(0, molblock_crn.GetNumAtoms()):
    pos = conf.GetAtomPosition(i)
    print(pos.x, pos.y, pos.z)
```

##### You can compute distances / geometry

### 6) Measure distances, angles, dihedrals

```python
import math

# distance between atom i and j
def dist(i, j):
    pi = conf.GetAtomPosition(i)
    pj = conf.GetAtomPosition(j)
    return math.sqrt((pi.x - pj.x)**2 + (pi.y - pj.y)**2 + (pi.z - pj.z)**2)

dist(0, 5)
```

##### RDKit can also do angles/dihedrals if you want.

### 7) Read residue/ chain info (PDB residue lables)

```python
# get the residue information for each atom. Here residues is a object.
residues = [atom.GetPDBResidueInfo() for atom in molblock_crn.GetAtoms()]

# from residues we can get the number and name of each residue
resinf = [(res.GetResidueNumber(), res.GetResidueName()) for res in residues]

# take a look of the last ten
resinf[-10:]
```

```python
residues = sorted(set(resinf))
len(residues)
```
